```python
import pickle
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib as mpl
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
%matplotlib inline
import seaborn as sns
```

```python
with open('model_cat.pkl', 'rb') as f:
    model1 = pickle.load(f)
```

```python
Xy_train_aa = pd.read_csv("X_train.csv")
Xy_test_aa = pd.read_csv("Y_train.csv")
```

```python
X_train = Xy_train_aa.iloc[:,:-1]
y_train = Xy_train_aa.iloc[:,-1]
X_test = Xy_test_aa.iloc[:,:-1]
y_test = Xy_test_aa.iloc[:,-1]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,classification_report, f1_score, recall_score
```

```python
y_pred = model1.predict(X_test)
```

```python
p_S = precision_score(y_test, y_pred)
a_S = accuracy_score(y_test, y_pred)
f1_S = f1_score(y_test, y_pred)
r_S = recall_score(y_test, y_pred)
c_M = confusion_matrix(y_test,y_pred)
```

```python
# my_classes = df['label'].unique()
my_classes = ['blues','classical','country','disco',\
              'hiphop','jazz','metal','pop','reggae','rock']
```

```python
fig, ax = plt.subplots()
sns.heatmap(c_M, annot=True, cmap='Blues', fmt='g', xticklabels=my_classes, yticklabels=my_classes, ax=ax)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
plt.show()
```
